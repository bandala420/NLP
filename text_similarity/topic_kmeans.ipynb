```python
# Daniel Bandala @ nov 2022
from __future__ import division
import numpy as np
from nltk import cluster
from nltk.cluster import euclidean_distance
import matplotlib.pyplot as plt
```

```python
vectors = np.array([np.array(f) for f in [[1, 1], [2, 1], [4, 3], [5, 4]]])
means = [[1, 1], [2, 1]]
k = 2
```

```python
clusterer = cluster.KMeansClusterer(k, euclidean_distance, initial_means=means)
clusters = clusterer.cluster(vectors, True)
centers = np.array(clusterer.means())
```

```python
print ('Clustered:', vectors)
print ('As:', clusters)
print ('Means:', clusterer.means())
```

```python
# Clasificar un nuevo vector
vector = np.array([3, 3])
print ('classify(%s):' % vector,)
print (clusterer.classify(vector))
```

```python
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'xx-large',
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
pylab.rcParams.update(params)
```

```python
plt.figure(figsize=(16,8))
plt.scatter(vectors[:,0],vectors[:,1], marker='*', linewidths=24)
plt.scatter(centers[:,0],centers[:,1], marker='x', linewidths=24)
plt.scatter(vector[0],vector[1], marker='o', linewidths=24)
plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
plt.legend(["Vectors","Centroids","New vector"])
plt.grid()
```
